```python
import cv2
import matplotlib.pyplot as plt
import glob
%matplotlib inline
```

```python
dataset = ['May26', 'Jul14'][1]
for i, fname in enumerate(sorted(glob.glob('jinpeng.{}/*.jpg'.format(dataset)))):
    img = cv2.imread(fname)
    h, w = img.shape[:2]
    if dataset == 'May26':
        crop = img[h//6:h*7//9,w//5:w*3//5]
    else:
        crop = img[h//6:h*5//6,w//5:w*4//5]
    plt.figure();plt.imshow(crop)
    cv2.imwrite('thumbs/{:03d}.jpg'.format(i), crop)
```
